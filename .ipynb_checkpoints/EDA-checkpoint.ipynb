```python
import pandas as pd
```

```python
df = pd.read_excel("./EDA.xlsx")
```

```python
df.columns
```

```python
df.head()
```

```python
def skillCorrection(skills):
    skills = str(skills).replace("\n", ",").replace(" ,", ",").replace("0", "Generic").strip()
    skills = skills.split(",")
    skills = {skill.upper().strip() for skill in skills} 
    
    return ", ".join(skills).strip(",")
```

```python
df['Primary Skill'].isnull().sum()
```

```python
df['Primary Skill'] = df['Primary Skill'].fillna("Generic")
df['Primary Skill'] = df['Primary Skill'].apply(skillCorrection)
df.head()
```

```python
skills_series = df['Primary Skill'].str.split(', ')
```

```python
skills_list = [skill for sublist in skills_series for skill in sublist]
```

```python
skill_counts = pd.Series(skills_list).value_counts()
```

```python
skill_counts
```

```python
skill_counts_filtered = skill_counts[skill_counts > 1]
skill_counts_filtered
```

```python
import matplotlib.pyplot as plt
```

```python

# Plot the graph
plt.figure(figsize=(10, 6))
skill_counts_filtered.plot(kind='bar', color='skyblue')
plt.title('Frequency of Skills')
plt.xlabel('Skill')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
